# Simple graph generation

## Example generations of star and grid graphs

```python
from PIL import Image

from simple_graph.graph_generation import star_graph, grid_graph

num_nodes = 12
num_edges = 14
num_rows = 3
num_cols = 4

img_star, edges_star = star_graph(num_nodes, num_edges, img_size=256, drift=20)
img_grid, edges_grid = grid_graph(num_rows, num_cols, num_edges, img_size=256, drift=20)

display(Image.fromarray(img_star), edges_star.__str__())
display(Image.fromarray(img_grid), edges_grid.__str__())
```

```python
from simple_graph.dataset import num_possible_star_graphs, num_possible_grid_graphs

print(num_possible_star_graphs(num_nodes, num_edges))
print(num_possible_grid_graphs(num_rows, num_cols, num_edges))
```
